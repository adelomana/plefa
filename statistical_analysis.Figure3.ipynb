```python
import pandas, numpy
import matplotlib, matplotlib.pyplot

import matplotlib, matplotlib.pyplot
matplotlib.rcParams.update({'font.size':20, 'font.family':'FreeSans', 'xtick.labelsize':20, 'ytick.labelsize':20, 'figure.figsize':(8, 4)})
```

# 0. user-defined variables

```python
data_file = '~/gd15/hi/research/plefa/data/figure3/partial_Day7_Results.csv'
figures_folder = '/Users/adrian/gd15/hi/research/plefa/results/figure3/'
```

```python
metabolites = ['EPA', '5-HEPE', '12-HEPE', '15-HEPE', '18-HEPE', 'DHA', '7-HDHA', '10-HDHA', '17-HDHA']
```

# 1. read information

```python
df = pandas.read_csv(data_file)
df
```

# 2. plot

```python
for metabolite in metabolites:
    kk = df[df['Group']=='K-K'][metabolite].values
    prima = df[df['Group']=='Primatrix'][metabolite].values
    
    kk = kk[numpy.isfinite(kk)]
    prima = prima[numpy.isfinite(prima)]

    print(metabolite)
    print(kk.shape, kk)
    print(prima.shape, prima)
        
    bp = matplotlib.pyplot.boxplot([kk], positions=[1], patch_artist=True, widths=2/3)
    matplotlib.pyplot.setp(bp['boxes'], color='black', lw=2, facecolor='purple')
    matplotlib.pyplot.setp(bp['medians'], color='black', lw=2) 
    matplotlib.pyplot.setp(bp['whiskers'], lw=2)
    matplotlib.pyplot.setp(bp['caps'], lw=2)

    
    bp = matplotlib.pyplot.boxplot([prima], positions=[2], patch_artist=True, widths=2/3)
    matplotlib.pyplot.setp(bp['boxes'], lw=2, color='black', facecolor='gold')
    matplotlib.pyplot.setp(bp['medians'], color='black', lw=2) 
    matplotlib.pyplot.setp(bp['whiskers'], lw=2)
    matplotlib.pyplot.setp(bp['caps'], lw=2)
    
    matplotlib.pyplot.grid(alpha=0.5, axis='y')
    matplotlib.pyplot.ylabel('log2 NSI')
    matplotlib.pyplot.title(metabolite)
    matplotlib.pyplot.tight_layout()
    
    matplotlib.pyplot.show()
    
    #figure_file = '{}{}.svg'.format(figures_folder, metabolite)
    #matplotlib.pyplot.savefig(figure_file)
    #matplotlib.pyplot.clf()
```
